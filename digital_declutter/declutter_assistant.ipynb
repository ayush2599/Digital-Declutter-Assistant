```python
%load_ext autoreload
%autoreload 2
import sys
import os
import nest_asyncio
nest_asyncio.apply()

# Add parent directory to path so we can import the 'digital_declutter' package
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from digital_declutter.agent import create_agent
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.genai import types

print("Libraries imported!")
```

```python
# Initialize the Agent
# Reloading agent module to pick up changes
import digital_declutter.agent
import importlib
importlib.reload(digital_declutter.agent)
from digital_declutter.agent import create_agent

agent = await create_agent(model_name="gemini-2.5-flash-lite")
session_service = InMemorySessionService()

# Define app_name FIRST
app_name = "declutter_app"

# Pass app_name to Runner
runner = Runner(agent=agent, app_name=app_name, session_service=session_service)

session_id = "session_1"
user_id = "user"

# Create session
import asyncio
async def init_session():
    await session_service.create_session(app_name=app_name, user_id=user_id, session_id=session_id)
    print("\nâœ… Agent initialized and ready!")

await init_session()
```

```python
# Chat Function
# Run this cell to define the chat function.

async def talk(user_message):
    print(f"You: {user_message}")
    message = types.Content(parts=[types.Part(text=user_message)])
    print("Agent is thinking...")
    
    async for event in runner.run_async(user_id=user_id, session_id=session_id, new_message=message):
        if event.is_final_response() and event.content:
            for part in event.content.parts:
                if hasattr(part, "text"):
                    print(f"Agent: {part.text}")
```

```python
# Usage Example
# Run this cell to send a message. Change the text below to whatever you want.
await talk("Clean my inbox")
```
